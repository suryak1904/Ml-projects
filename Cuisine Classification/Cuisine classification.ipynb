```python
df_encoded=pd.get_dummies(df,columns=[
    'Cuisines', 'Currency', 'Has Table booking',
       'Has Online delivery', 'Is delivering now', 'Switch to order menu',
       'Price range', 'Rating color','Rating text'])
df_encoded
```

```python
user_preferences={
    'Cuisines': input("enter the Cuisine:"),
    'Price range':int(input("enter the price range:")),
    'Has Table booking': input("Has table booking:"),
    'Has Online delivery': input("has online delivery:"),
    'Is delivering now': input("is delivering now:")
}
user_profile = pd.DataFrame([user_preferences])
user_profile_encoded = pd.get_dummies(user_profile)
user_profile_encoded = user_profile_encoded.reindex(columns=df_encoded.columns, fill_value=0)
user_profile_encoded = user_profile_encoded.drop(columns=[
    'Restaurant ID', 'Restaurant Name', 'Country Code', 'City', 'Address',
    'Locality', 'Locality Verbose', 'Longitude', 'Latitude', 'Average Cost for two',
    'Votes', 'Aggregate rating'
])
similarities = cosine_similarity(df_encoded.drop(columns=[
    'Restaurant ID', 'Restaurant Name', 'Country Code', 'City', 'Address',
    'Locality', 'Locality Verbose', 'Longitude', 'Latitude', 'Average Cost for two',
    'Votes', 'Aggregate rating'
]), user_profile_encoded)
similarities = similarities.flatten()
df['similarity'] = similarities
recommendations = df.sort_values(by='similarity', ascending=False).head(1)
print(recommendations[['Restaurant Name', 'Cuisines', 'Price range', 'Aggregate rating', 'similarity']])
```
